```python
def lista(n):
    list_a=list(range(0, n, 2))
    list_b=list(range(0, n, 10))
    list_c=list(range(0, n, 5))
    new_list=[0]*n
    if len(list_a)==len(list_b)==len(list_c):
        print("длина равна")
    for i in range(len(list_a)):
        for j in range (len(list_b)):
            if list_a[i]==list_b[j]:
                for k in range(len(list_c)):
                    if list_c[k]==list_b[j]:
                        new_list[k]=list_c[k]
    return new_list
```

```python
new=lista(1000)
print(new)
#убрать нули
```

```python
class specification_valves:
    def __init__(self, name, obj, col, ceh,work):
        self.name = name #название
        self.obj = obj #чертеж
        self.col= col # количество
        self.ceh= ceh # цех
        self.work= work # трудоемкость н/часы
    def __str__(self): 
        return  "name:% s obj:% s col:% s цех:% s  work:% s" % (self.name, self.obj, self.col, self.ceh, self.work) 
    
    
corpus = specification_valves('corpus', '20-20',100,6,200)
lopatca =  specification_valves('lopatca', '50-50',1000,6,700)
roof =  specification_valves('roof','30-30', 200,9,500)

List_sort = [corpus,lopatca,roof]

List_sort.sort(key=lambda x: x.ceh)

for i in range(len(List_sort)):
    print(List_sort[i])
```
